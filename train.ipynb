Important imports.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

import xgboost as xgb

from src.utils import load_json_data, train_test_split_data, plot_cross_val_results, perform_kfold_cross_validation
from src.data_preparation import preprocess_data
from src.model import Model
```

Load the raw data file.

```python
data = load_json_data(r'data\raw\data.json')
```

```python
data
```

Preprocess the data.

```python
df = preprocess_data(data)
```

```python
df.head()
```

Split the data into train and test data sets.

```python
import warnings
warnings.filterwarnings('ignore', category=UserWarning)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split_data(df)
```

We want to train 5 classifier model and compare their performances. Initialize the classifier models.

```python
models = {
    'SVC': SVC(probability=True),
    'LogisticRegression': LogisticRegression(),
    'DecisionTree': DecisionTreeClassifier(),
    'RandomForest': RandomForestClassifier(),
    'XGBoost': xgb.XGBClassifier()
}
```

Perform the K-Fold cross-validation.

```python
# Perform k-fold cross-validation
print('\nPerforming k-fold cross-validation...')
print('='*37)
cross_val_results = perform_kfold_cross_validation(X_train, y_train, models, n_splits=10)
```

Plot the cross-validation results.

```python
plot_cross_val_results(cross_val_results)
```

Train the models and test them on the test data set, to check the performance metrics.

```python
metrics_data = []

# Train and evaluate models on the test set
for name, model_classifier in models.items():
    print(f'\nTraining and evaluating {name}...')
    model = Model(model_classifier)
    model.train(X_train, y_train)
    metrics_data.append({name: model.evaluate(X_test, y_test)})
    model.save(r'models\\', name)
```

Compare the performance metrics in a plot.

```python
# Extract metric names and values
metric_names = ['Accuracy', 'Precision', 'Recall', 'F1-score']
model_names = list(models.keys())
metric_values = [[list(data.values())[0][metric] for metric in metric_names] for data in metrics_data]

# Plotting
num_models = len(model_names)
num_metrics = len(metric_names)
bar_width = 0.15
index = np.arange(num_models)

fig, ax = plt.subplots(figsize=(25, 6))

for i in range(num_metrics):
    ax.bar(index + i * bar_width, [values[i] for values in metric_values], bar_width, label=metric_names[i])

ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Model Performance Metrics')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(model_names)
ax.legend()

plt.tight_layout()
plt.show()
plt.savefig('images\train_metrics.png', dpi=300)
```
